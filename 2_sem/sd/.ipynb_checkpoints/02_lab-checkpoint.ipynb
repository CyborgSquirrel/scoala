```python
x = var('x')
y = function('y')(x)
eqd = diff(y, x) == 2*x*(1+y^2)
show(eqd)
sol = desolve(eqd, y)
show(solve(sol, y))
```

```python
x = var('x')
y = function('y')(x)
eqd = (x^2-1)*diff(y, x) + 2*x*y^2 == 0
sol = desolve(eqd, y)
show(solve(sol, y))
```

```python
x = var('x')
y = function('y')(x)
eqd = 2*x^2*diff(y) == x^2+y^2
sol = desolve(eqd, y)
show(solve(sol, y))
```

```python
x = var('x')
y = function('y')(x)
eqd = diff(y) == -(x+y)/y
sol = desolve(eqd, y)
show(sol)
```

```python
x = var('x')
y = function('y')(x)
eqd = diff(y, x, 2) + diff(y, x) == sin(x) + cos(x)
sol = desolve(eqd, y)
show(sol)
```

```python
x = var('x')
y = function('y')(x)
eqd = diff(y, x, 2) - diff(y, x) == e^(2*x)
sol = desolve(eqd, y)
show(sol)
```

# 2

```python
x = var('x')
y = function('y')(x)
eqd = diff(y) == 1 + y^2
sol = desolve(eqd, y, ics =[0,1])
sol = solve(sol, y)[0]
show(sol)
plot(sol.rhs(), -10, 10, detect_poles=True, ymin=-50, ymax=50)
```

```python
x = var('x')
y = function('y')(x)
eqd = diff(y) == 1/(1-x^2)*y+1+x
sol = desolve(eqd, y, ics=[2,0])
show(sol)
plot(sol, x, 1.1, 100)
```

```python
x = var('x')
y = function('y')(x)
eqd = diff(y) - 2*y == -x^2
sol = desolve(eqd, y, ics=[0,1/4])
show(sol)
plot(sol, -10, 10)
```

```python
x = var('x')
y = function('y')(x)
eqd = diff(y, x, 2) - 5*diff(y) + 4*y == 0
sol = desolve(eqd, y, ics=[0,5,8])
show(sol)
plot(sol, -10, 1)
```

```python
x = var('x')
y = function('y')(x)
eqd = diff(y, x, 2) - 4*diff(y) + 5*y == 2*x^2*e^x
sol = desolve(eqd, y, ics=[0,2,3])
show(sol)
plot(sol, 11, 14)
```

```python
x = var('x')
y = function('y')(x)
eqd = diff(y, x, 2) + 4*y == 4*(sin(2*x) + cos(2*x))
sol = desolve(eqd, y, ics=[pi, 2*pi, 2*pi])
show(sol)
plot(sol, 0, 10)
```

# 3

```python
k = var('k')
x = var('x')
y = function('y')(x)
eqd = diff(y) + k/x*y == x^3
sol(x,k,_C) = desolve(eqd, y, ivar=x)
show(sol)
plot([sol(x, 1, C) for C in [-5..5]], -10, 10, ymax=10, ymin=-10)
```

```python
k = var('k')
x = var('x')
y = function('y')(x)
eqd = diff(y) + k/x*y == x^3
sol(x,k) = desolve(eqd, y, ivar=x, ics=[1,0])
show(sol(x,1))
animate([plot(sol(x,k*2), -10, 10, ymin=-10, ymax=10, detect_poles=True) for k in [1..20]])
```

# 4

```python
a, b = var('a, b')
x = var('x')
y = function('y')(x)
eqd = diff(y) == y + a*x + b
sol(x, a, b) = desolve(eqd, y, ivar=x, ics=[0,1])
val = solve([sol(1, a, b) == -5+4*e, sol(2, a, b) == -7+4*e^2], a, b)
show(val)
plot(sol(x, val[0][0].rhs(), val[0][1].rhs()), 0, 10)
```

```python
a = var('a')
x = var('x')
y = function('y')(x)
deq = diff(y, x, 2) - diff(y) - 2*y == 0
sol(x, a) = desolve(deq, y, ics=[0, a, 2])
show(sol)
limit(sol(x, -2), x=+infinity)
```
