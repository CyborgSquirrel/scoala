# 3

(a)

Malthus

```python
r,x0 = var('r,x0')
t = var('t')
x_m = function('x_m')(t)
sol_m(t,x0,r) = desolve(diff(x_m) == r*x_m, x_m, ivar=t, ics=[0,x0])
show(sol_m)
```

Verhulst

```python
r0,K,x0 = var('r0,K,x0')
C = var('C')
t = var('t')
x_v = function('x_v')(t)
sol_v = desolve(diff(x_v) == r0*x_v*(1-x_v/K), x_v, ivar=t, ics=[0,x0])
show(sol_v)

sol_v = sol_v.log_simplify()
sol_v(t,x0,r0,K) = solve(sol_v, x_v)[0].rhs()
show(sol_v)
```

(b)

```python
plot([sol_m(t,10,r) for r in [-1, -0.5, 0, 0.5, 1]], 0, 2)
```

```python
plot([sol_v(t,10,r,100) for r in [-1, -0.5, 0, 0.5, 1]], 0, 10)
```

(c)

```python
sol_m2(t,r) = sol_m(t,25*10^3,r)
r = solve(sol_m2(2,r) == 30*10^3, r)[1]
r = r.rhs()
p = sol_m2(5,r)
show(numerical_approx(p))
```

(d)

```python
show(sol_v)
sol_v2(t,r0,K) = sol_v(t,20*10^3,r0,K)
sol_K = solve([sol_v2(2,r0,K) == 40*10^3], K)[0]
sol_r0 = solve([(sol_v2(3,r0,K) == 50*10^3).substitute(K = sol_K.rhs())], r0)[1]
sol_K = sol_K.substitute(r0 = sol_r0.rhs())

sol_v3(t) = sol_v2(t,sol_r0.rhs(),sol_K.rhs())
show(numerical_approx(sol_v3(7)))

show("K=", numerical_approx(sol_K.rhs()))
show("r0=", numerical_approx(sol_r0.rhs()))

# @interact
# def _(a=(0,100),b=(0,100)):
show(plot(sol_v3(t), 0, 20))
```

# 4

(a)

```python
v_0,R,g = var('v_0,R,g')
x = var('x')
v = function('v')(x)
sol = desolve(v*diff(v,x) == -(g*R^2)/(x+R)^2, v, ivar=x, ics=[0,v_0])
sol = solve(sol, v)
sol(x,v_0,R,g) = sol[1].rhs()
show(sol)
```

(b)

```python
sol_2(x) = sol(x,50,6371,9.81)
sol_2(75)
```

(c)

```python
sol_c = solve(sol_2(x) == 0, x)[0]
sol_c = numerical_approx(sol_c.rhs())
show(sol_c)
```

```python
plot(sol_2(x), 0, 130)
```

(d)

```python
v_0f(x,R,g) = solve(sol(x,v_0,R,g) == 0, v_0)[1].rhs()
show(v_0f)
```

```python
evel(R,g) = limit(v_0f(x,R,g), x=infinity)
show(evel)
show(numerical_approx(evel(6371*10^3,9.81)))
```
