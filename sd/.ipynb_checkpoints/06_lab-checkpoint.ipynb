```python
s = var('s')
f = function('f')(s)
f(s) = s^2 - 2*s
fd = diff(f,s)

sols = solve(f(s) == 0, s)
show(sols)

for sol in sols:
    show(sol, '\tf\'(s)=', fd(sol.rhs()))
```

- local asimptotic stabil in $s=0$
- instabil in $s=2$

(c)

```python
s = var('s')
f = function('f')(s)
f(s) = sin(s)
fd(s) = diff(f,s)

sols = solve(f(s) == 0, s, to_poly_solve='force')
show(sols)

n = var('n')
with assuming(n, 'integer'):
    k = 2*n
    show(fd(k*pi))
    k = 2*n+1
    show(fd(k*pi))
```

- instabil in $s=2k\pi$
- asimptotic local stabil in $s=(2k+1)\pi$
