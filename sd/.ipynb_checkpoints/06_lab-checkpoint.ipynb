```python
s = var('s')
f = function('f')(s)
f(s) = s^2 - 2*s
fd = diff(f,s)

sols = solve(f(s) == 0, s)
show(sols)

for sol in sols:
    show(sol, '\tf\'(s)=', fd(sol.rhs()))
```

- local asimptotic stabil in $s=0$
- instabil in $s=2$

```python
plot_slope_field(f(s), (t,-5,5), (s,-3,3), headaxislength=3, headlength=4)
```

(b)

```python
s = var('s')
f = function('f')(s)
f(s) = s * (s-1) * (s-2)
fd = diff(f,s)

sols = solve(f(s) == 0, s)
show(sols)

for sol in sols:
    show(sol, '\tf\'(s)=', fd(sol.rhs()))
```

- instabil in $s=0$
- local asimptotic stabil in $s=1$
- instabil in $s=2$

```python
plot_slope_field(f(s), (t,-5,5), (s,-3,3), headaxislength=3, headlength=4)
```

(c)

```python
s = var('s')
f = function('f')(s)
f(s) = sin(s)
fd(s) = diff(f,s)

sols = solve(f(s) == 0, s, to_poly_solve='force')
show(sols)

n = var('n')
with assuming(n, 'integer'):
    k = 2*n
    show(fd(k*pi))
    k = 2*n+1
    show(fd(k*pi))
```

- instabil in $s=2k\pi$
- asimptotic local stabil in $s=(2k+1)\pi$

```python
t=var('t')
x=function('x')(t)
y=function('y')(t)

f1(x,y)=2*x+y
f2(x,y)=x+2*y

n=sqrt(f1(x,y)^2+f2(x,y)^2)
g=plot_vector_field([f1(x,y)/n,f2(x,y)/n], [x,-5,5], [y,-5,5], aspect_ratio=1)
for x0,y0 in [ [1,0],[0,1],[1,1],[-1,-1] ]:
    P=desolve_system_rk4([f1(x,y),f2(x,y)], [x,y], ics=[0,x0,y0], ivar=t, end_points=[-5,5], step=0.01)
    g+=list_plot([[j,k] for i,j,k in P], plotjoined=True, color='blue')
g.show(xmin=-5,ymin=-5,xmax=5,ymax=5)
```

```python
A=matrix([[2,1],[1,2]])
show(A.eigenvalues())
```

### 3

(a)

```python
x,y,t=var('x,y,t')
f1(x,y)=y
f2(x,y)=x*(1-x^2)+y
sols=solve([f1(x,y)==0, f2(x,y)==0], [x,y])
show(sols)
```

```python
J=jacobian((f1(x,y), f2(x,y)), (x,y))
show(J)
for sol in sols:
    A=J(x=sol[0].rhs(), y=sol[1].rhs())
    show(sol)
    show(A.eigenvalues())
```

(-1, 0) este instabil, de tip focus

(1, 0) este instabil, de tip focus

(0, 0) este instabil, de tip focus

```python
n=sqrt(f1*f1+f2*f2)
g=plot_vector_field([f1(x,y)/n, f2(x,y)/n], [-5,5], [-5,5])
g+=point([-1,0])
g+=point([1,0])
g+=point([0,0])
show(g)
```

(b)

```python
x,y,t=var('x,y,t')
f1(x,y)=-2*x+y+2
f2(x,y)=x*y

sols=solve([f1(x,y)==0, f2(x,y)==0], [x,y])
J=jacobian((f1(x,y), f2(x,y)), (x,y))
show(J)
for sol in sols:
    show(sol)
    show(J(x=sol[0].rhs(), y=sol[1].rhs()).eigenvalues())
```

(1, 0) este instabil, de tip sa

(0, -2) este asimptotic stabil

```python
n=sqrt(f1*f1+f2*f2)
g=plot_vector_field([f1(x,y)/n, f2(x,y)/n], [-5,5], [-5,5])
g+=point([1,0])
g+=point([0,-2])
show(g)
```

(c)

```python
x,y,t=var('x,y,t')
f1(x,y)=x^2-y^2
f2(x,y)=x*y-1

sols=solve([f1(x,y)==0, f2(x,y)==0], [x,y])
J=jacobian((f1(x,y), f2(x,y)), (x,y))
for sol in sols:
    if sol[0].rhs().is_real() and sol[1].rhs().is_real():
        show(sol)
        show(J(x=sol[0].rhs(), y=sol[1].rhs()).eigenvalues())
```

(-1, -1) este local asimptotic stabil

(1, 1) este instabil, de tip focus

```python
n=sqrt(f1*f1+f2*f2)
g=plot_vector_field([f1(x,y)/n, f2(x,y)/n], [-5,5], [-5,5])
g+=point([-1,-1])
g+=point([1,1])
show(g)
```

(a)

```python
x,y,t=var('x,y,t')
f1(x,y)=2*x-1.2*x*y
f2(x,y)=-y+0.9*x*y
```

(b)

```python
J=jacobian((f1,f2), (x,y))
sols=solve([f1(x,y)==0, f2(x,y)==0], [x,y])
for sol in sols:
    show(sol)
    show(J(x=sol[0].rhs(), y=sol[1].rhs()).eigenvalues())
```

(0,0) este instabil, de tip sa

(10/9, 5/3) este instabil, de tip nod

(c)

```python
n=sqrt(f1*f1+f2*f2)
g=plot_vector_field([f1(x,y)/n, f2(x,y)/n], [-5,5], [-5,5], aspect_ratio=1)
P=desolve_system_rk4([f1(x,y), f2(x,y)], [x,y], ics=[0,0.5,2], ivar=t, end_points=[-5,5], step=0.01)
g+=list_plot([[j,k] for i,j,k in P], plotjoined=True)
g+=point([0,0])
g+=point([10/9,5/3])
show(g)
```
