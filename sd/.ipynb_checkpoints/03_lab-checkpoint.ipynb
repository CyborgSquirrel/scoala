```python
t=var('t')
x=function('x')(t)
y=function('y')(t)
syst=[diff(x,t)==x+4*y,diff(y,t)==x+y]
sol=desolve_system(syst, [x,y], [0,1,2])
show(sol)
plot([sol[0].rhs(), sol[1].rhs()])
```

```python
t=var('t')
x=function('x')(t)
y=function('y')(t)
syst=[diff(x,t)==x-y+t-1, diff(y,t)==-2*x+4*y+cos(t)]
sol=desolve_system(syst, [x,y], [0,0,1])
show(sol)
plot([sol[0].rhs(), sol[1].rhs()])
```

```python
t=var('t')
x=function('x')(t)
y=function('y')(t)
syst=[diff(x,t)==x+2*y+e^(-t), diff(y,t)==-2*x+y+1]
sol=desolve_system(syst, [x,y], [0,0,1])
show(sol)
plot([sol[0].rhs(), sol[1].rhs()])
```

```python
t=var('t')
x=function('x')(t)
y=function('y')(t)
z=function('z')(t)
syst=[diff(x,t)==-x+3*y+3*z+27*t^2, diff(y,t)==2*x-2*y-5*z+3*t, diff(z,t)==-2*x+3*y+6*z+3]
sol=desolve_system(syst, [x,y,z], [0,50,-30,-26])
show(sol)
plot([sol[0].rhs(), sol[1].rhs(), sol[2].rhs()])
```

# 3

```python
f1(u,v)=u+v
f2(u,v)=-2*u+4*v
d(u,v)=sqrt(f1(u,v)^2+f2(u,v)^2)
g=plot_vector_field([f1(u,v)/d(u,v), f2(u,v)/d(u,v)], [u,-10,10], [v,-10,10])

t=var('t')
x=function('x')(t)
y=function('y')(t)
syst=[diff(x,t)==x+y, diff(y,t)==-2*x+4*y]

for con,col in zip([[0,3,0],[0,0,3],[0,-3,0],[0,0,-3]], rainbow(4)):
    sol=desolve_system(syst, [x,y], con)
    show(sol)
    sol_x(t)=sol[0].rhs()
    show(limit(sol_x(t), t=+infinity))
    sol_y(t)=sol[1].rhs()
    show(limit(sol_y(t), t=+infinity))
    g += parametric_plot((sol_x(t), sol_y(t)), (t, -10, 10), color=col)

g.show(xmin=-5,xmax=5,ymin=-5,ymax=5)
```
