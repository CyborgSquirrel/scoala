```python
var('t')
var('s')
f(t,s) = 1 - (s*cos(4*t)*cos(t)) / sqrt(1-s^2*cos(4*t)^2*sin(t)^2)

def p(s):
    return parametric_plot((f(t-pi/2, s), f(t, s)), (t, 0, 2*pi))

show(p(0.5))

g = p(0.1)
for i in srange(0.2, 1, 0.1):
    g += p(i)
show(g)
```
