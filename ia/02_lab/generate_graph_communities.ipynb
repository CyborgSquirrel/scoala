```python
import networkx
import random

edges_inside = 60
edges_outside = 7

community_sizes = [ 15, 8, 4 ]
community_chances = [ community_size*(community_size-1)/2 for community_size in community_sizes ]
print(community_chances)
nodes_len = sum(community_sizes)

graph = networkx.Graph()

# for node_index in range(nodes_len):
#     graph.add_node(node_index)

# add edges inside
for _ in range(edges_inside):
    success = False
    while not success:
        communities = list(range(len(community_sizes)))
        community_chances_aux = list(community_chances)
        
        community_index = random.choices(communities, weights=community_chances_aux)[0]
        community_start = sum(community_sizes[0:community_index])
        
        node_1 = (
              community_start
            + random.randint(0, community_sizes[community_index]-1)
        )
        node_2 = (
              community_start
            + random.randint(0, community_sizes[community_index]-1)
        )
                
        if node_1 != node_2 and not graph.has_edge(node_1, node_2):
            graph.add_edge(node_1, node_2)
            success = True
        # else:
        #     print(node_1, node_2)

# add edges outside
for _ in range(edges_outside):
    success = False
    while not success:
        communities = list(range(len(community_sizes)))
        community_chances_aux = list(community_chances)
        
        community_1_index = random.choices(communities, weights=community_chances_aux)[0]
        
        del communities[community_1_index]
        del community_chances_aux[community_1_index]
        
        community_2_index = random.choices(communities, weights=community_chances_aux)[0]
        
        community_1_start = sum(community_sizes[0:community_1_index])
        community_2_start = sum(community_sizes[0:community_2_index])
        
        community_1_node = (
              community_1_start
            + random.randint(0, community_sizes[community_1_index]-1)
        )
        community_2_node = (
              community_2_start
            + random.randint(0, community_sizes[community_2_index]-1)
        )
        
        # print(community_1_node, community_2_node)
        if not graph.has_edge(community_1_node, community_2_node):
            graph.add_edge(community_1_node, community_2_node)
            success = True

networkx.draw(graph)
```

```python
networkx.write_gml(graph, "./my-2.gml")
```
